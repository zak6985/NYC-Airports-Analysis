```python
Carrier = flights_data.carrier.unique()
Carrier.sort()
print(Carrier)
```

```python
Total_no_flights = []
Total_airmiles = []
Total_airtime = []
Airmiles_less_than_1000 = []
Airmiles_less_than_2000 = []
Airmiles_more_than_2000 = []
Total_Delay = []
Total_Delay_L1000 = []
Total_delay_M1000_L2000 = []
Total_Delay_M2000 = []
Average_Distance = []
Average_Speed = []
Speed_L150 = []
Speed_M150_L300 = []
Speed_M300 = []
Destination = []
for i in Carrier:
    flights_data = flights_data[~pd.isnull(flights_data.air_time)] # removed all cancelled flights
    Airmiles = flights_data.distance[flights_data.carrier == i]
    flights_data_carrier = flights_data[flights_data.carrier == i]
    Airtime = flights_data.air_time[flights_data.carrier == i]
    Airmiles_L1000 = Airmiles[Airmiles<1000]
    Airmiles_L2000 = Airmiles[Airmiles<2000]
    Airmiles_M2000 = Airmiles[Airmiles>2000]
    Airmiles_M1000 = Airmiles_L2000.count()- Airmiles_L1000.count()
    Delay = flights_data.arr_delay[flights_data.carrier == i]
    Delay_L1000 = flights_data_carrier.arr_delay[flights_data_carrier.distance<1000]
    Delay_M1000 = flights_data_carrier.arr_delay[flights_data_carrier.distance>1000]
    Delay_M2000 = flights_data_carrier.arr_delay[flights_data_carrier.distance>2000]
    Delay_B_1000_2000 = Delay_M1000.sum()-Delay_M2000.sum()
    Speed = (Airmiles/Airtime)*60
    Speed_150 = Speed[Speed<150].count()
    Speed_B_150_300 = Speed[Speed<300].count()-Speed[Speed<150].count()
    Speed_300 = Speed[Speed>300].count()
    Total_airmiles.append(Airmiles.sum())
    Total_airtime.append(Airtime.sum())
    Airmiles_less_than_1000.append(Airmiles_L1000.count())
    Airmiles_less_than_2000.append(Airmiles_M1000)
    Airmiles_more_than_2000.append(Airmiles_M2000.count())
    Total_no_flights.append(Airmiles.count())
    Total_Delay.append(Delay.sum())
    Total_Delay_L1000.append(Delay_L1000.sum())
    Total_delay_M1000_L2000.append(Delay_B_1000_2000)
    Total_Delay_M2000.append(Delay_M2000.sum())
    Average_Speed.append(Speed.mean())
    Average_Distance.append(Airmiles.mean())
    Speed_M150_L300.append(Speed_B_150_300)
    Speed_L150.append(Speed_150)
    Speed_M300.append(Speed_300)
```
